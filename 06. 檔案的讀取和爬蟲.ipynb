在 Jupyter Notebook 中, 我們可以顯示各種東西...

```python
from IPython.display import display, HTML, Markdown, Image, YouTubeVideo
```

### 先來 `Markdown`

```python
md_text = "## 你好!"
```

用 `Markdown` 包起來。

```python
md01 = Markdown(md_text)
```

再 display 顯示出來!

```python
display(md01)
```

這和直接在 Jupyter Notebook 筆記的 `Markdown` 有什麼不同? 答案是現在可以放在程式中...

```python
for i in range(3):
    display(md01)
```

數學符號也是可以的!

```python
md_text = '''
## 來試試數學符號

$$L = \int_a^b f(x) \, dx$$
'''
```

```python
display(Markdown(md_text))
```

於是可以出個印出來很漂亮的數學題目!

```python
from random import randint

a = randint(1, 9)
b = randint(2, 9)
c = randint(1, 9)
d = randint(2, 9)

display(Markdown('## 請回答下列問題'))

display(Markdown('$$\dfrac{%d}{%d} + \dfrac{%d}{%d} = ?$$' % (a,b,c,d)))
```

### HTML

HTML 是寫網頁的程式語言, 想學可以參考

[W3Schools HTML 教學](https://www.w3schools.com/html/default.asp)

```python
html_txt = "<h2>你好!</h2>"
```

```python
egg = HTML(html_txt)
```

```python
display(egg)
```

那看起來比 `Markdown` 麻煩一些耶! 用 HTML 到底是有什好處? 原來 HTML 自由度更高, 可以做的變化更多!

```python
html_txt = '<h2 style="color:red;">你好!</h2>'
```

```python
display(HTML(html_txt))
```

```python
html_txt = '上了這門課, <s>就會和我一樣厲害</s>'
```

```python
display(HTML(html_txt))
```
